### Handling Outliers

#### identifying and removing outliers using IQR

```python
Q1 = dataY["Gender"].quantile(0.25)
Q3 = dataY["Gender"].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
lower_limit, upper_limit
```

```python
data_between = dataY[dataY["Gender"].between(lower_limit, upper_limit, inclusive='left')]
data_between
```

```python
data_no_outliers = dataY[dataY["Gender"].between(lower_limit, upper_limit, inclusive='neither')]
data_no_outliers
```
