# Create a DataFrame from the Data obtained from the API get request

## Drop some unnecessary columns

```python
import pandas as pd
df = pd.DataFrame(standings)
df.drop(['league_id', 'season_id', 'group_id', 'team_id', 'group_name', 'stage_name', 'stage_id', 'competition_id'], axis=1, inplace=True)
print(df)
```

```python
df.info()
```

## Convert data from some columns to numerical type

```python
df_copy = df.copy()
df_copy['rank'] = df_copy['rank'].astype(int)
df_copy['points'] = df_copy['points'].astype(int)
df_copy['matches'] = df_copy['matches'].astype(int)
df_copy['goal_diff'] = df_copy['goal_diff'].astype(int)
df_copy['goals_scored'] = df_copy['goals_scored'].astype(int)
df_copy['goals_conceded'] = df_copy['goals_conceded'].astype(int)
df_copy['lost'] = df_copy['lost'].astype(int)
df_copy['won'] = df_copy['won'].astype(int)
df_copy['drawn'] = df_copy['drawn'].astype(int)

df_copy.info()
```

## Sort by point in descending order

```python
df_copy.sort_values(by='points', ascending=False, inplace=True)
print(df_copy)
```

## Sort by Goal difference

```python
df_copy.sort_values(by='goal_diff', ascending=False, inplace=True)
print(df_copy[['rank','name', 'points','goal_diff']])
```

## Sort by amount of games won

```python
df_copy.sort_values(by='won', ascending=False, inplace=True)
print(df_copy[['rank','name', 'points','won']])
```
